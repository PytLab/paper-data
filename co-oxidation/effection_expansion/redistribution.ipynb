# 扩展supercell

```python
nodes, edges = [], []
```

## 创建nodes 和 edges

### top位相关

```python
from catplot.grid_components.nodes import Node2D
from catplot.grid_components.edges import Edge2D
```

```python
top = Node2D([0.0, 0.0], size=800, color="#2A6A9C")
```

```python
t1 = Node2D([0.0, 1.0])
t2 = Node2D([1.0, 0.0])
```

```python
nodes.append(top)
```

```python
e1 = Edge2D(top, t1, width=5)
e2 = Edge2D(top, t2, width=5)
```

```python
edges.extend([e1, e2])
```

### bridge相关

```python
bridge1 = Node2D([0.0, 0.5], style="s", size=600, color="#5A5A5A", alpha=0.6)
bridge2 = Node2D([0.5, 0.0], style="s", size=600, color="#5A5A5A", alpha=0.6)
```

```python
b1 = bridge1.clone([0.5, 0.5])
b2 = bridge2.clone([0.5, 0.5])
```

```python
nodes.extend([bridge1, bridge2])
```

```python
e1 = Edge2D(bridge1, b1)
e2 = Edge2D(bridge1, bridge2)
e3 = Edge2D(bridge2, b2)
e4 = Edge2D(b1, b2)
```

```python
edges.extend([e1, e2, e3, e4])
```

### hollow位

```python
h = Node2D([0.5, 0.5], style="h", size=700, color="#5A5A5A", alpha=0.3)
```

```python
nodes.append(h)
```

## 绘制

```python
from catplot.grid_components.grid_canvas import Grid2DCanvas
```

```python
from catplot.grid_components.supercell import SuperCell2D
```

```python
supercell = SuperCell2D(nodes, edges)
```

## 将此supercell进行扩展

```python
expanded_supercell = supercell.expand(4, 4)
```

## 绘制

```python
canvas_big = Grid2DCanvas(figsize=(30, 20), dpi=60)
```

```python
canvas_big.add_supercell(expanded_supercell)
```

```python
canvas_big.draw()
canvas_big.figure
```

## 创建空位species node

```python
empty = Node2D([0.0, 0.0], size=3000, 
               color="none", edgecolor="#000000", line_width="2", line_style="dashed", zorder=3, alpha=0.7)
```

## 创建O2分子

```python
o1 = Node2D([0.0, 0.0], size=3000, color="#B95D60", zorder=4, alpha=0.5)
```

```python
o2 = o1.clone([1.0, 0.0])
```

```python
bond = Edge2D(o1, o2, width=15, alpha=0.5)
```

```python
o2 = SuperCell2D([o1, o2], [bond])
```

## 创建CO分子

```python
co = Node2D([0.0, 0.0], size=3000, color="#A472C8", zorder=4, alpha=0.5)
```

## 向画布中添加原子

```python
n0 = empty.clone([1.0, 0.5])
n1 = o1.clone([3.5, 2.0])
n2 = co.clone([1.0, 1.5])
n3 = co.clone([2.0, 2.5])
n4 = co.clone([2.5, 2.0])
n5 = o1.clone([1.5, 2.0])
n6n7 = o2.clone([2.0, 1.0])
n8 = o1.clone([1.0, 2.5])
n9 = empty.clone([2.0, 1.5])
n10 = empty.clone([2.0, 3.5])
```

```python
canvas_big.add_nodes([n1, n2, n3, n4, n5, n8])
canvas_big.add_supercell(n6n7)
```

```python
canvas_big.redraw()
canvas_big.figure
```

## 创建area supercell

```python
# O 空位
o_empty = Node2D([0.0, 0.0], size=3000, color="none", edgecolor="#F44E47", line_width=3, alpha=0.7, line_style="dashed")
```

```python
# 0th
n2 = o_empty.clone([1.0, 2.5])
n3 = o_empty.clone([2.0, 2.5])
n4 = o_empty.clone([1.0, 1.5])
n5 = o_empty.clone([2.0, 1.5])

nn_0_emtpy = [n2, n3, n4, n5]

# 1st
o_emtpy_1 = Node2D([0.0, 0.0], size=3000, color="none", edgecolor="#F44E47", line_width=3, alpha=0.5, line_style="dashed")
nn_positions = [[0.5, 1.0], [0.5, 1.5], [0.5, 2.0], [0.5, 2.5], [0.5, 3.0], [1.0, 3.0], [1.5, 3.0], [2.0, 3.0],
                [2.5, 3.0], [2.5, 2.5], [2.5, 2.0], [2.5, 1.5], [2.5, 1.0], [2.0, 1.0], [1.5, 1.0], [1.0, 1.0]]
nn_1_emtpy = []
for p in nn_positions:
    i = o_emtpy_1.clone(p)
    nn_1_emtpy.append(i)
```

```python
from catplot.grid_components.supercell import SuperCell2D
all_nodes = nn_0_emtpy + nn_1_emtpy
supercell = SuperCell2D(all_nodes, [])
```

## 克隆supercell

```python
s1 = supercell.clone([-0.5, -0.5])
s1.set_nodes_attr("edgecolor", "#242424")
s1.set_nodes_attr("size", 4500)
s2 = supercell.clone([0.5, 0.5])
s2.set_nodes_attr("edgecolor", "#26BB5A")
s2.set_nodes_attr("size", 6000)
s3 = supercell.clone([1.0, 0.0])
```

```python
canvas_big.add_supercells([s1, s2, s3])
```

```python
canvas_big.redraw()
canvas_big.figure
```
