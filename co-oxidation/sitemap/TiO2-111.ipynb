# 绘制TiO2 (110) 面

```python
from catplot.grid_components.nodes import Node2D
from catplot.grid_components.edges import Edge2D
nodes, edges = [], []
```

```python
ti = Node2D([0.0, 0.0], size=800, color="#6D6E70", alpha=1.0, zorder=1)
o = Node2D([0.0, 0.0], size=800, color="#E35858", alpha=1.0, zorder=1)
```

## Ti nodes

```python
t1 = ti.clone()
t2 = ti.clone([0.5, 1.0])
```

```python
nodes.extend([t1, t2])
```

## O nodes

```python
from math import sqrt
o1 = o.clone([0.0, 1.0])
o2 = o.clone([0.5, 0.5/sqrt(3)])
o3 = o.clone([0.5, 2.-0.5/sqrt(3)])
```

```python
nodes.extend([o1, o2, o3])
```

## 辅助nodes

```python
n1 = Node2D([0.0, 2.0])
n2 = Node2D([1.0, 2.0])
n3 = Node2D([1.0, 1.0])
n4 = Node2D([1.0, 0.0])
```

## Edges

```python
e1 = Edge2D(t1, n4, width=5)
e2 = Edge2D(o2, t1, width=5)
e3 = Edge2D(o2, n4, width=5)
e4 = Edge2D(t2, n3, width=5)
e5 = Edge2D(t2, o1, width=5)
e6 = Edge2D(o3, t2, width=5)
e9 = Edge2D(o2, t2, width=5)
e7 = Edge2D(o3, n1, width=5)
e8 = Edge2D(o3, n2, width=5)
```

```python
edges.extend([e1, e2, e3, e4, e5, e6, e7, e8, e9])
```

## 绘制

```python
from catplot.grid_components.supercell import SuperCell2D
```

```python
supercell = SuperCell2D(nodes, edges)
```

```python
from catplot.grid_components.grid_canvas import Grid2DCanvas
```

```python
canvas = Grid2DCanvas(margin_ratio=0.3)
```

```python
canvas.add_supercell(supercell)
```

```python
canvas.draw()
canvas.figure
```

## 扩展

```python
expanded_supercell = supercell.expand(3, 2, cell_vectors=[[1.0, 0.0], [0.0, 2.0]])
```

```python
canvas_big = Grid2DCanvas(figsize=(30, 20), dpi=60)
canvas_big.add_supercell(expanded_supercell)
```

```python
canvas_big.draw()
canvas_big.figure
```

## 创建空位点

```python
empty = Node2D([0.0, 0.0], size=3000, 
               color="none", edgecolor="#006DD1", line_width=6, line_style="dashed", zorder=3, alpha=0.5)
oads = Node2D([0.0, 0.0], size=3000, color="#B95D60", line_width=2, zorder=3, alpha=0.7)
```

```python
em1 = oads.clone([1.0, 1.0])
em2 = empty.clone([1.0, 2.0])
em3 = empty.clone([2.0, 2.0])
em4 = empty.clone([2.0, 1.0])
em5 = empty.clone([1.5, 1.0])
em6 = empty.clone([1.5, 2.0-0.5/sqrt(3)], edgecolor="#009A61")
em7 = empty.clone([1.5, 2.0+0.5/sqrt(3)])
```

```python
canvas_big.add_nodes([em1, em2, em3, em4, em5, em6, em7])
```

```python
canvas_big.redraw()
canvas_big.figure
```
