# 扩展supercell

```python
nodes, edges = [], []
```

## 创建nodes 和 edges

### top位相关

```python
from catplot.grid_components.nodes import Node2D
from catplot.grid_components.edges import Edge2D
```

```python
top = Node2D([0.0, 0.0], size=800, color="#2A6A9C")
```

```python
t1 = Node2D([0.0, 1.0])
t2 = Node2D([1.0, 0.0])
```

```python
nodes.append(top)
```

```python
e1 = Edge2D(top, t1, width=5)
e2 = Edge2D(top, t2, width=5)
```

```python
edges.extend([e1, e2])
```

### bridge相关

```python
bridge1 = Node2D([0.0, 0.5], style="s", size=600, color="#5A5A5A", alpha=0.6)
bridge2 = Node2D([0.5, 0.0], style="s", size=600, color="#5A5A5A", alpha=0.6)
```

```python
b1 = bridge1.clone([0.5, 0.5])
b2 = bridge2.clone([0.5, 0.5])
```

```python
nodes.extend([bridge1, bridge2])
```

```python
e1 = Edge2D(bridge1, b1)
e2 = Edge2D(bridge1, bridge2)
e3 = Edge2D(bridge2, b2)
e4 = Edge2D(b1, b2)
```

```python
edges.extend([e1, e2, e3, e4])
```

### hollow位

```python
h = Node2D([0.5, 0.5], style="h", size=700, color="#5A5A5A", alpha=0.3)
```

```python
nodes.append(h)
```

## 绘制

```python
from catplot.grid_components.grid_canvas import Grid2DCanvas
```

```python
from catplot.grid_components.supercell import SuperCell2D
```

```python
supercell = SuperCell2D(nodes, edges)
```

## 将此supercell进行扩展

```python
expanded_supercell = supercell.expand(4, 4)
```

## 绘制

```python
canvas_big = Grid2DCanvas(figsize=(30, 20), dpi=60)
```

```python
canvas_big.add_supercell(expanded_supercell)
```

```python
canvas_big.draw()
canvas_big.figure
```

## 创建空位species node

```python
empty = Node2D([0.0, 0.0], size=3000, 
               color="none", edgecolor="#009A61", line_width=6, line_style="dashed", zorder=3, alpha=0.7)
```

## 创建O2分子

```python
o1 = Node2D([0.0, 0.0], size=3000, color="#B95D60", zorder=4, alpha=0.5)
```

```python
o2 = o1.clone([1.0, 0.0])
```

```python
bond = Edge2D(o1, o2, width=15, alpha=0.5)
```

```python
o2 = SuperCell2D([o1, o2], [bond])
```

## 创建CO分子

```python
co = Node2D([0.0, 0.0], size=3000, color="#A472C8", zorder=4, alpha=0.5)
```

## 创建一个❌

```python
wrong = Node2D([0.0, 0.0], size=2000, color="#EA0000", zorder=4, alpha=0.5, style="X")
```

## 向画布中添加原子

```python
b = o1.clone([1.5, 2.0])
w1 = wrong.clone([1.0, 2.0])
w2 = wrong.clone([1.5, 2.5])
w3 = wrong.clone([2.0, 2.0])
w4 = wrong.clone([1.5, 1.5])
e1 = empty.clone([1.0, 1.5])
e2 = empty.clone([1.0, 2.5])
e3 = empty.clone([2.0, 2.5])
e4 = empty.clone([2.0, 1.5])
canvas_big.add_nodes([b, w1, w2, w3, w4, e1, e2, e3, e4])
```

```python
canvas_big.redraw()
canvas_big.figure
```

## 删除红色nodes

```python
canvas_big.remove(b, w1, w2, w3, w4)
```

```python
e5 = empty.clone([1, 2])
e6 = empty.clone([1.5, 1.5])
e7 = empty.clone([2, 2])
e8 = empty.clone([1.5, 2.5])
canvas_big.add_nodes([e5, e6, e7, e8])
```

```python
canvas_big.redraw()
canvas_big.figure
```
