```python
%matplotlib inline
```

```python
from catplot.ep_components.ep_lines import ElementaryLine
from catplot.ep_components.ep_canvas import EPCanvas
```

```python
canvas = EPCanvas(figsize=(9, 6), dpi=200)
```

```python
l1 = ElementaryLine([0.0, -2.09], color="#5898D8")
```

```python
canvas.add_line(l1)
```

```python
canvas.draw()
canvas.figure
```

```python
l2 = ElementaryLine([0.0, 1.18, -0.59],color="#B57C7A")
```

```python
canvas.add_line(l2)
```

```python
canvas.draw()
canvas.figure
```

```python
l2.translate(-1.5, "y")
```

```python
canvas.redraw()
canvas.figure
```

```python
l1.shadow_color
```

```python
l1.shadow_depth
```

```python
l1.shadow_depth = 5
l2.shadow_depth = 5
```

```python
canvas.redraw()
canvas.figure
```
